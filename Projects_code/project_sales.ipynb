```python
df_salesorder_1=pd.read_csv('D:\python_promming\Project\salesorder.csv',usecols=['ORDERID','TOTALAMOUNT','CUSTOMERID'],index_col='CUSTOMERID')
```

```python
df_orderdetails_1 = pd.read_csv('D:\python_promming\Project\orderdetails.csv',usecols=['ORDER_DETAILS_ID','QUANTITY','ORDERID','PRODUCTID' ],index_col='PRODUCTID')
```

```python
df_orderdetails_1 
```

```python
df_product_1 = pd.read_csv('D:\python_promming\Project\product.csv',usecols=['PRODUCTID','PRODUCTNAME'],index_col='PRODUCTID')
```

```python
merged_df_1 = pd.merge(df_customer_1, df_salesorder_1, on='CUSTOMERID', how='inner')
```

```python
merged_df_1
```

```python
merged_df_2 = pd.merge(df_orderdetails_1, df_product_1, on='PRODUCTID', how='inner')
```

```python
final_merge=pd.merge(merged_df_1,merged_df_2, on ='ORDERID')
```

```python
final_merge.set_index('ORDERID', inplace=True)
```

```python
final_merge
```
